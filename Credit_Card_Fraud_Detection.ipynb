Importing the dependencies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
1 # loading the dataset to a Pandas DataFrame
credit_card_data = pd.read_csv('/content/creditcard.csv')
```

```python
# The datasets
credit_card_data.head()
```

```python
credit_card_data.tail()
```

```python
# dataset info
credit_card_data.info()
```

```python
# distribution of correct transactions & fraud transactions
credit_card_data['Class'].value_counts()
```

0 --> Correct tansactions

1 --> Fraud Transactions

```python
# separating the data for checking
correct = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
print(correct.shape)
print(fraud.shape)
```

```python
# statistical  measure of the data
correct.Amount.describe()
```

```python
fraud.Amount.describe()
```

Under-Sampling:
Build a sample dataset containing similar distribution of noormal transactions and fradulant Transactions

Number of fraudlent Transaction  -->38

```python
correct_sample = correct.sample(n=38)
```

```python
new_dataset =pd.concat([correct_sample, fraud], axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

```python
X = new_dataset.drop(columns='Class',axis=1)
Y = new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2,stratify=Y, random_state =3 )
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
model = LogisticRegression()
```

```python
# training the model
model.fit(X_train,Y_train)
```

```python
#model evaluation
# accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy on training data : ',training_data_accuracy)
```

```python
# accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction,Y_test)
```

```python
print('Accuracy score on test Dta : ',test_data_accuracy )
```
